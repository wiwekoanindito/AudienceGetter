```python
import pandas as pd
import re
import os
```

```python
COLUMNS = [
    "Name",
    "ID",
    "Type",
    "Source",
    "Display size",
    "YouTube users",
    "Mobile app",
    "CTV devices",
    "Display line items",  
]

def parse_dv360_text_file(filepath):
    """Parse DV360 copied text where each record has 9 lines including '—'."""
    with open(filepath, "r", encoding="utf-8") as f:
        # Keep separator lines ("—"), only remove blanks
        lines = [line.strip() for line in f if line.strip()]

    # Break into chunks of 9 lines per record
    records = [lines[i:i+9] for i in range(0, len(lines), 9)]

    df = pd.DataFrame(records, columns=COLUMNS)
    return df
```

```python
for file in os.listdir('Raw Data'):
    df = parse_dv360_text_file(file)
    df.to_excel("clean/audience.xlsx", index=False)
    df.to_csv("clean/audience.csv", index=False, encoding="utf-8-sig")
    print(f"✅ Cleaned {file}")
```
